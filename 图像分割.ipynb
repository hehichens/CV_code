# 分水岭算法函数watershed

```python
# -*- coding: cp936 -*-
import cv2
import numpy as np

# Step1. 加载图像
img = cv2.imread("C:/Users/hichens/Desktop/CV_code/images/fruit.png")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(10, 8))
plt.subplot(231)
plt.title("original image")
plt.imshow(img, cmap="gray")
plt.axis("off")

# Step2.阈值分割，将图像分为黑白两部分
ret, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
plt.subplot(232)
plt.title("thresh image")
plt.imshow(thresh, cmap="gray")
plt.axis("off")


# Step3. 对图像进行“开运算”，先腐蚀再膨胀
kernel = np.ones((3, 3), np.uint8)
opening = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel, iterations=2)
plt.subplot(233)
plt.title("opening image")
plt.imshow(opening, cmap="gray")
plt.axis("off")


# Step4. 对“开运算”的结果进行膨胀，得到大部分都是背景的区域
sure_bg = cv2.dilate(opening, kernel, iterations=3)
plt.subplot(234)
plt.title("sure_bg image")
plt.imshow(sure_bg, cmap="gray")
plt.axis("off")


# Step5.通过distanceTransform获取前景区域
dist_transform = cv2.distanceTransform(opening, cv2.DIST_L2, 5)
ret, sure_fg = cv2.threshold(dist_transform, 0.6 * dist_transform.max(), 255, 0)
plt.subplot(235)
plt.title("sure_bg image")
plt.imshow(sure_bg, cmap="gray")
plt.axis("off")


# Step6. sure_bg与sure_fg相减,得到既有前景又有背景的重合区域
sure_fg = np.uint8(sure_fg)
unknow = cv2.subtract(sure_bg, sure_fg)

# Step7. 连通区域处理
ret, markers = cv2.connectedComponents(sure_fg)
markers = markers + 1
markers[unknow==255] = 0

# Step8.分水岭算法
markers = cv2.watershed(img, markers)
img[markers == -1] = [255, 0, 0] #红色标记分水岭
plt.subplot(236)
plt.title("dst image")
plt.imshow(img, cmap="gray")
plt.axis("off")

plt.show()
```

# GrabCut算法函数grabCut

```python
# -*- coding: cp936 -*-

import cv2
import numpy as np

# Step1. 加载图像
img = cv2.imread("C:/Users/hichens/Desktop/CV_code/images/fruit.png")

# Step2. 创建掩模、背景图和前景图
mask = np.zeros(img.shape[:2], np.uint8)# 创建大小相同的掩模
bgdModel = np.zeros((1,65), np.float64)# 创建背景图像
fgdModel = np.zeros((1,65), np.float64)# 创建前景图像

# Step3. 初始化矩形区域
# 这个矩形必须完全包含前景
rect = (50,0,450,333) #格式为（x, y, w, h）

# Step4. GrubCut算法，迭代5次
# mask的取值为0,1,2,3
cv2.grabCut(img, mask, rect, bgdModel, fgdModel, 5, cv2.GC_INIT_WITH_RECT) # 迭代5次

# Step5. mask中，值为2和0的统一转化为0, 1和3转化为1
mask2 = np.where((mask == 2) | (mask == 0), 0, 1).astype('uint8')
img = img * mask2[:,:,np.newaxis] # np.newaxis 插入一个新维度，相当于将二维矩阵扩充为三维

plt.imshow(img, cmap="gray")
plt.axis("off")
plt.show()
```
